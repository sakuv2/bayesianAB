```python
from scipy import stats
import numpy as np


@np.vectorize
def normal_sampling(s2):
    return stats.norm.rvs(size=1, loc=0, scale=1 / (1 * s2))

s2s = np.array([0.5, 1, 1.2])
xs = normal_sampling(s2s)

np.stack([xs, s2s]).T
```
